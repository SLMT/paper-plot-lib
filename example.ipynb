```python
import paper_plot as pt
```

# Basic Usages

## To Create a Line Plot

```python
pt.lines_from_csv('examples/example_data.csv').draw()
```

## To Set Labels

```python
pt.lines_from_csv('examples/example_data.csv') \
    .x_label('Time (seconds)') \
    .x_label_size(15) \
    .y_label('Performance') \
    .y_label_size(15) \
    .draw()
```

## To Show Legend

```python
pt.lines_from_csv('examples/example_data.csv') \
    .show_legend() \
    .draw()
```

## To Pull the Legend Out

```python
pt.lines_from_csv('examples/example_data.csv') \
    .legend_out() \
    .draw()
```

## To Change Colors of the Lines

```python
pt.lines_from_csv('examples/example_data.csv') \
    .colors(['#008828', '#121259', '#df5349']) \
    .legend_out() \
    .draw()
```

## To Add Markers to the Lines

To check marker options, please see [the document of matplotlib](https://matplotlib.org/3.2.2/api/markers_api.html).

```python
pt.lines_from_csv('examples/example_data.csv') \
    .markers(['o', '^', 's']) \
    .legend_out() \
    .draw()
```

## To Change Line Styles

To check style options, please see [the document of matplotlib](https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html).

```python
pt.lines_from_csv('examples/example_data.csv') \
    .line_styles(['--', ':', '-.']) \
    .legend_out() \
    .draw()
```
